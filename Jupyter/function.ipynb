```python
#   f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
#1.  
#2.  ,    
#3.  ,    
#4.  
#5.  
#6. ,   f > 0
#7. ,   f < 0
from sympy import*
from sympy.plotting import plot
from matplotlib import*
init_printing()
x = Symbol('x')
f = -12*x*x*x*x*sin(cos(x))-18*x*x*x+5*x*x+10*x-30
plot(f,(x,-100,100),title = ' ')
```

```python
print(f'  : {limit(f, x, -oo)}')
print(f'  : {limit(f, x, oo)}')
print(f'    Y: [0 ,{f.subs(x, 0)}]')
print(f' f(x) = {f},        ')
```

```python
print(f' 1- : {f.diff(x)}')
```

```python
df = f.diff(x)
print('  : \n(df < 0) -   \n(df > 0) -  ')
plot(f,(x,-25,25),line_color='red',title=' f(x)')
plot(df,(x,-25,25),line_color='blue',title=' df(x)')

```
